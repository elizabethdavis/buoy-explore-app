```python
from bs4 import BeautifulSoup
import requests
```

url = "https://www.ndbc.noaa.gov/activestations.xml"

```python
# get the data
xml = requests.get(url)
```

```python
# parse the data
soup = BeautifulSoup(xml.content, 'lxml')
print(soup)
```

```python
import pandas as pd

station_list = soup.find_all('station')

for x in station_list:
    id = x.get('id')
    name = x.get('name')
    met = x.get('met')
    
    print("ID: " + id)
    print("Name: " + name)
    print("Met Data: " + met)
```
